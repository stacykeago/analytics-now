```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
model= linear_model.LinearRegression()
```

```python
data= pd.read_csv("Salary_dataset.csv")
data.head()
```

```python
data.columns
```

```python
x_data =np.array(data ['YearsExperience']).reshape(-1,1)
y_data =np.array(data ['Salary']).reshape(-1,1)
```

```python
#print(x_data.shape)
#print(y_data.shape)
```

```python
model= linear_model.LinearRegression()
model.fit(x_data,y_data)
plt.scatter(x_data,y_data)
```

```python
# split into train and test
from sklearn.model_selection import train_test_split

x_test, x_train, y_test, y_train =train_test_split(x_data,y_data,test_size=0.2, random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# predict the salary based on the years of experience
y_pred= model.predict(np.array([6]).reshape(-1,1))
print(y_pred)
```

```python
y_pred = model.predict(x_test)
r2score = model.score(x_test,y_test)
print(r2score)
```
